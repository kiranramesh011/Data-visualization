# Axes, Labels, Legend, Grid

```python
days=[1,2,3,4,5,6,7]
max_t = [52,51,50,48,49,46,47]
min_t = [36,42,40,44,43,35,35]
avg_t = [48,46,45,46,46,40,41]
```

```python
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t, label='Max')
plt.plot(days,min_t, label = 'Min')
plt.plot(days,avg_t, label = 'Avg')
plt.legend(loc = 'lower left', shadow= False, fontsize="large")
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t, label='Max')
plt.plot(days,min_t, label = 'Min')
plt.plot(days,avg_t, label = 'Avg')
plt.legend(loc = 'best', shadow= False, fontsize="large")
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t, label='Max')
plt.plot(days,min_t, label = 'Min')
plt.plot(days,avg_t, label = 'Avg')
plt.legend(loc = 'best', shadow= True, fontsize="large")
plt.grid()
```

```python
plt.barh(xpos,revenue, label="Revenue")

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```

# HISTOGRAM
- In histograms, x axis contains a variable and y axis will be a frequency of that variable
We have a sample data of blood sugar level of different patients, we will try to plot number of patients by blood range and try to figure out how many patients are normal, pre-diabetic and diabetic

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.8) # by default number of bins is set to 10
```

```python
plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar,rwidth=0.5,bins=4)
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g')
```

# Mutiple data samples in a histogram

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.legend()
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,histtype='step')
```

# horizontal orientation

```python
plt.xlabel("Number Of Patients")
plt.ylabel("Sugar Level")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, orientation='horizontal')
```

```python
## Scatter plot
sns.scatterplot(x='sepal_length',y='sepal_width',data=df,hue='species')
plt.show()

#In the Seaborn library, the "hue" parameter is used to visualize data from different categories in a single plot.
```

```python
## Bar plot
##A barplot is basically used to aggregate the categorical data 
#according to some methods and by default its the mean. 
##It can also be understood as a visualization of the group by action. 
#To use this plot we choose a categorical column for 
##the x axis and a numerical column for the y axis and we see that it 
# creates a plot taking a mean per categorical column. 

sns.barplot(x='species',y='petal_length', data=df)
plt.show()
```

```python
## Count Plot
##A countplot basically counts the categories and returns a 
# count of their occurrences. 
sns.countplot(x='species', data=df)
plt.show()
```

```python
sns.catplot(x='species',data=df)
plt.show()
```

```python
sns.catplot(x='petal_length',y='species',data=df)
plt.show()
```

```python
## Box plot
##A boxplot is sometimes known as the box and whisker plot.It shows the 
#distribution of the quantitative data 
## that represents the comparisons between variables. boxplot shows the 
#quartiles of the dataset while the whiskers extend to
## show the rest of the distribution i.e. the dots indicating the 
#presence  of outliers.
sns.boxplot(x='sepal_width',data=df)
plt.show()
```

# The formula to identify outliers using the Interquartile Range (IQR) is: outlier = any data point below (Q1 - 1.5 * IQR) or above (Q3 + 1.5 * IQR), where Q1 is the first quartile, Q3 is the third quartile, and IQR is the interquartile range (calculated as Q3 - Q1).

```python
sns.boxplot(x='species', y='sepal_width', data=df)
plt.show()
# Checking outliers in each category with respect to sepal width.
#one of the major use of boxplot is to find the outlier 
```

```python
sns.boxplot(x='species', y='petal_width', data=df)
plt.title("box plot for iris")
plt.show()
```

```python
## Violinplot
## It is similar to the boxplot except that it provides a higher,
## more advanced visualization and uses the kernel 
##density estimation to give a better description about the 
## data distribution.

sns.violinplot(x='species', y='sepal_width', data=df)
plt.show
```

```python
##Stripplot
## It basically creates a scatter plot based on the category. 
# It is created using the stripplot()
sns.stripplot(x='species', y='petal_width', data=df)
plt.show()
```

```python
## Histogram
## A histogram is basically used to represent data provided in a form of 
# some groups.
## It is accurate method for the graphical representation of numerical data 
## distribution
sns.histplot(x='petal_width', data=df)
plt.show
```

```python
sns.histplot(x='petal_width',data=df,hue='species',kde=True)
plt.show()
#kde is approximation of histogram to density curve.
```

```python
## Distplot
## Distplot is used basically for univariant set of observations
# and visualizes it through a histogram 
## i.e. only one observation and hence we choose one particular 
#column of the dataset
sns.distplot(df['petal_width'],hist=True,color="r")
plt.show()
## density means the probability of given point to lie on the curve.
```

```python
##Pairplot
##Pairplot represents pairwise relation across the entire dataframe and supports an 

## additional  argument called hue for categorical separation. 
#What it does basically is create a jointplot between every possible numerical 
##column and takes a while if the dataframe is really huge

sns.pairplot(data=df,hue='species')
plt.show()
```
